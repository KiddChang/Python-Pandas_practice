```python
data['pick_date'] = data['pick'].apply(lambda x: pd.Timestamp(x, unit='s', tz='Asia/Singapore').date())
```

```python
data.head()
```

```python
data['first_date'] = data['1st_deliver_attempt'].apply(lambda x: pd.Timestamp(x, unit='s', tz='Asia/Singapore').date())
data['second_date'] = data['2nd_deliver_attempt'].apply(lambda x: pd.Timestamp(x, unit='s', tz='Asia/Singapore').date())
```

```python
data.head()
```

```python
holidays = [
    '2020-03-01',
    '2020-03-08',
    '2020-03-15',
    '2020-03-22',
    '2020-03-25',
    '2020-03-29',
    '2020-03-30',
    '2020-03-31',
    '2020-04-05'
]

h_days = [datetime.strptime(i, '%Y-%m-%d').date() for i in holidays]
```

```python
h_days
```

```python
data.head()
```

```python
h_days[1]
```

```python
data.loc[(data['pick_date'] <= h_days[1]) & (data['first_date'] >= h_days[1])]
```

```python
data['h_first'] = 0
data['h_second'] = 0
for i in h_days:
    data.loc[(data['pick_date'] <= i) & (data['first_date'] >= i), 'h_first'] += 1
    data.loc[(data['first_date'] <= i) & (data['second_date'] >= i), 'h_second'] += 1
```

```python
h_days
```

```python
data.head()
```

```python
data['h_first'] = data['h_first'].apply(lambda x: timedelta(days=x))
data['h_second'] = data['h_second'].apply(lambda x: timedelta(days=x))
```

```python
data.head()
```

```python
data['final_first'] = data['first_date'] - data['pick_date'] - data['h_first']
data['final_second'] = data['second_date'] - data['first_date'] - data['h_second']
```

```python
data.head()
```

```python
[i for i in data['buyeraddress'][:5]]
```

```python
data['b_add'] = data['buyeraddress'].apply(lambda x: x.split(' ')[-1])
data['s_add'] = data['selleraddress'].apply(lambda x: x.split(' ')[-1])
```
