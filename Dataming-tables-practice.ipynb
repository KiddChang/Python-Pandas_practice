## merge

```python
df1= pd.DataFrame({'id':[8, 7, 10, 17],
                  'name':['Tiffany', 'Joanne', 'Alex', 'Phoebe'],
                  'age':[38, 37, 42,25]})
df1
```

```python
df2= pd.DataFrame({'ID':[20, 8,10 ,17],
                  'name':['Ryan','Tiffany','Alex','Phoebe'],
                  'sex':['M','F','N','F']})
df2
```

```python
pd.merge(df1,df2) #只會將相同欄位的相同資料內容合併。
```

```python
pd.merge(df1,df2, on='name')
```

```python
pd.merge(df1,df2, how='inner') #與上頭一樣，所以預設的合併方式就為inner
```

```python
pd.merge(df1,df2,how='left') #取左合併 left joining
```

```python
pd.merge(df1,df2,how='right') #取右合併 right joining
```

```python
pd.merge(df1, df2,how='outer') #outer joining
```

```python
pd.merge(df1,df2, left_on='id',right_on='ID') #用左id 與右Id合併 相同資料的留下
```

## concatenate

```python
df3= pd.DataFrame({'id':[8, 7, 10, 17],
                  'name':['Tiffany', 'Joanne', 'Alex', 'Phoebe'],
                  'age':[38, 37, 42,25]})
df3
```

```python
df4 = pd.DataFrame({'id':[20],
                   'name':['Ryan'],
                   'age':[24]})
df4
```

```python
df = pd.concat([df3,df4]) #直接合併表格，但index 仍為0
df
```

```python
df = pd.concat([df3,df4],ignore_index=True) #用ignore_index=True  可自動排序加入後的index
df
```

```python
df5 = pd.DataFrame({'sex':['F','F','M','M','F']})
df5
```

```python
df = pd.concat([df,df5], axis =1) #預設值是併row 如果要併column 要加 axis=1
df
```

## pivot table

```python
import pandas as pd
iris = pd.read_csv('iris.csv')
```

```python
iris
```

```python
iris.Species.unique() #三種species
```

```python
iris['pl'] = pd.qcut(iris['PetalLengthCm'], 3, labels=['short', 'medium','long']) #用qcut 將petallengthcm 分成三個等級
```

```python
iris['pw'] = pd.qcut(iris['PetalWidthCm'], 3, labels=['short', 'medium','long']) #用qcut 將petalwidthcm 分成三個等級
```

```python
iris
```

```python
pd.pivot_table(iris, index=['Species']) #pivot_table內只會有數字  預設值為平均值
```

```python
pd.pivot_table(iris, index='Species', aggfunc='count') #看index 的數目
```

```python
pd.pivot_table(iris, index=['Species','pl'], aggfunc='count') #取兩個index  pl 可再細分資料分佈
```

```python
pd.pivot_table(iris, index=['Species','pl'], aggfunc=['min','max','mean']) #也可加入複數的功能
```

```python
pd.pivot_table(iris, index='Species', columns='pl',values='PetalWidthCm')
```

```python
pd.pivot_table(iris, index='Species', columns='pl',values=['PetalWidthCm','SepalWidthCm'])
```
