```python
import seaborn as sns
import pandas as pd
```

```python
iris=pd.read_csv('iris.csv')
iris
```

## scatter plot

```python
sns.scatterplot(data=iris, x='PetalLengthCm', y='PetalWidthCm');
```

```python
sns.scatterplot(data=iris, x='PetalLengthCm',y='PetalWidthCm', hue='Species')
```

```python
sns.lineplot(data= iris, x= 'PetalLengthCm', y= 'PetalWidthCm' ,color ='r');
```

```python
sns.lineplot(data = iris, x= 'PetalLengthCm', y= 'PetalWidthCm', ci=None); #就沒有透明的了
```

```python
sns.lineplot(data = iris, x='PetalLengthCm', y='PetalWidthCm', hue ='Species' ,ci=None);
```

## bar graph

```python
sns.barplot(data=iris, x='Species', y= 'PetalLengthCm')
```

```python
sns.barplot(data=iris, x='Species', y= 'PetalLengthCm', ci=None)
```

```python
sns.countplot(data=iris, x='Species') #算次數的
```

## box plot

```python
sns.boxplot(data =iris, y='PetalLengthCm')
```

```python
sns.boxplot(data = iris ,x='Species', y='PetalLengthCm')
```

## pair plot

```python
sns.pairplot(iris) #還不知資料長怎樣時，先看的
```

```python
sns.pairplot(iris, hue='Species')
```

```python
sns.pairplot(iris, hue='Species', diag_kind='hist') #改變圖形
```
