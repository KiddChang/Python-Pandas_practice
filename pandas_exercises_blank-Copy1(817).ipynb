```python
def ohe_embarked(data):
    embarked = pd.get_dummies(df['Embarked'])
    data = pd.concat([data, embarked], axis=1)
    return data
```

```python
def ohe_title(data):
    title = pd.get_dummies(df['Title'])
    data = pd.concat([data, title], axis=1)
    return data
```
