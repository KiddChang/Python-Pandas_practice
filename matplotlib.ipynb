## bar chart

```python
x = ['apple', 'banana', 'orange', 'peach']
x_count = [5,6,3,10]
```

```python
plt.bar(x, x_count, color='c');
```

```python
plt.pie(x_count, labels=x);
```

```python
plt.pie(x_count, labels=x, explode=[0.1, 0, 0.3,0]);
```
