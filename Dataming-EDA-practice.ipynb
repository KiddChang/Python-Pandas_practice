## function &mapping

```python
x = pd.DataFrame([['Zoe',32, np.nan, 59 , 'banana'],
                 ['Sam',np.nan,'M', np.nan,'peach'],
                ['Danny',46,'M',81, 'apple']], columns=['name','age','sex','grade','fruit'])
x
```

```python
x=x.drop_duplicates()
```

```python
x
```

```python
x = x.fillna({'sex':'F','grade':65})
x
```

```python
sex ={'M':1,'F':0}
```

```python
x['sex'].map(sex) #map 有點類似取代  將x表的sex 代入上頭，字典的資料。
```

```python
x['sex_map']=x['sex'].map(sex)  #新增一個sex_map欄位
```

```python
x
```

```python
def letter_grade(grade):
    if grade < 60:
        letter ='F'
    elif grade < 70:
        letter = 'D'
    elif grade < 80:
        letter = 'C'
    elif grade < 90:
        letter ='B'
    else:
        letter ='A'
    return letter
```

```python
x['grade_map']= x['grade'].map(letter_grade) #map也可以帶入function
```

```python
x
```

```python
x['grade_apply']= x['grade'].apply(letter_grade) #用apply功能也可以
```

## bin

```python
grades = [45, 82, 56, 77, 90, 69, 72, 79, 85, 95, 62, 65]
```

```python
pd.cut(grades, 5) #把上述的資料範圍(最大到最小)製成5等區間
```

```python
df = pd.DataFrame({'grades':grades, 'cut':pd.cut(grades, 5)}) #( 不包含   , ] 包含
df
```

```python
df.iloc[0,1]
```

```python
df.iloc[1,1]
```

```python
df.iloc[0,1] < df.iloc[1,1]
```

```python
df['cut'].value_counts()  #看切成的區間有幾筆資料(看落在哪些區域)
```

```python
df['letter_cut']= pd.cut(grades, 5, labels=['F','D','C','B','A']) #自動分好等級
df
```
